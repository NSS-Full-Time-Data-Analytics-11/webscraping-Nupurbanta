```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import time
```

```python
URL='https://realpython.github.io/fake-jobs/'
```

```python
response=requests.get(URL)
response.status_code
```

```python
soup = BS(response.text)
```

```python
soup
```

```python
title_tags=soup.find('h2')
```

```python
title_list = [title.text for title in title_tags]
title_list
```

```python
company_tags = soup.findAll('h3')
```

```python
company_list = [company.text for company in company_tags]
company_list
```

```python
p_list= soup.findAll('p', attrs={'class' : 'location'})
```

```python
location_list=[location.text.strip() for location in p_list]
len(location_list)
```

```python
date_tags=soup.findAll('time')
```

```python
date_list=[date.text for date in date_tags] 
date_list
```

```python
date_tags2=soup.findAll('p', attrs={'class' : 'is-small'})
```

```python
date_list2=[date.text.strip() for date in date_tags2] 
date_list2
```

```python
jobs_df=pd.DataFrame({'title':title_list,'company': company_list,'location':location_list,'date':date_list})
jobs_df
```

```python
a_tags=soup.findAll('a')
```

```python
for tag in a_tags:
    print(tag['href'])
```
